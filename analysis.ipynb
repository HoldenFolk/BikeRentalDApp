```python
import matplotlib.pyplot as plt

# Sample data
x = [1995, 2009, 2013, 2024]
y1 = [100, 50, 30, 20]
y2 = [20, 60, 90, 100]

# Create a figure and a set of subplots
fig, ax1 = plt.subplots()

# First line (privacy)
ax1.plot(x, y1, 'g-', label='user\nprivacy')  # Green solid line

#ax1.set_ylabel('privacy level', color='g', )  # Green color for the privacy label
ax1.set_yticks([0, 20, 40, 60, 80, 100])
ax1.set_yticklabels(['None', 'Very Low', 'Low', 'Medium', 'High', 'Very High'])


# Second line (technology)
ax2 = ax1.twinx()
ax2.plot(x, y2, 'b-', label='technological\nsophistication')  # Blue solid line
#ax2.set_ylabel('technological\nsophistication', color='b')  # Blue color for the technology label
ax2.set_yticks([])


# Add scatter points
ax1.scatter(x, y1, color='g')
ax2.scatter(x, y2, color='b')

special_years = {1995: 'Bycyklen, 1995', 2009: 'Launch of BIXI, 2009', 2013: 'Introduction of\nBIXI app, 2013', 2024: 'Present'}
ax1.set_xticks(x)
ax1.set_xticklabels([special_years.get(year, '') for year in x], rotation=60)

# Title and legend

# Title and legend
plt.title('Technological evolutions of Bike sharing services and their consequences on user privacy.')
ax1.legend(loc='center left')
ax2.legend(loc='center right')

# Add labels for each point
# for i, (year, privacy, technology) in enumerate(zip(x, y1, y2)):
#     ax1.text(year, privacy, f'({privacy}, {technology})', ha='right', va='bottom')

# Show plot
plt.show()
```
